```python
class employee():
    Company = 'mast Company'
    
    def __init__(self,name,age,salary):
        self.name = name
        self.age = age
        self.salary=salary
    
    def welcome(self):
        print('Welcome {} to mast company at the age of {} with salary of {} thousand'.format(self.name,self.age,self.salary))
              
    def firstmonth(self):
        print('Well done!!',self.name)
        print('Your age is',self.age)
        print('your salary is',self.salary)

emp = employee('nikesh',25,30)
emp.welcome()
emp.firstmonth()

```

```python
#method verriding


class employee():
    def add(self,a,b):
        print('sum of a and b is',a+b)
        
        
class team(employee):
    def add(self,a,b,c):
        print('sum of a and b and c  is',a+b+c)
        
        
emp = employee()
emp.add(20,30)
Batch=team()
Batch.add(10,20,10)
```
