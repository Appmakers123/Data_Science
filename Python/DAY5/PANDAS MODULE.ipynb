```python
import numpy as np
import pandas as pd
```

```python
#merging ad joininga
```

```python
left = pd.DataFrame({'Key':['one','two'],'Value':[10,20]})
```

```python
right = pd.DataFrame({'Key':['one','two','Three'],'Value':[30,40,50],'Decription':['sample1','sample2','sample3']})
print(left)
print('\n\n',right)
```

```python
pd.merge(left, right,on ='Key',how ='left')
```

```python
pd.merge(left, right,on ='Key',how ='right')
```

```python
pd.merge(left, right,on ='Key',how ='outer')
```

```python
pd.merge(left, right,on ='Key',how ='inner') #common in both the tables
```
