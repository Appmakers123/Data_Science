```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
#Simple line chart

Height = [150,160,165,185]
Weight = [70,80,90,100]

plt.plot(Height,Weight) # x ,y
plt.show()
```

```python
plt.plot(Height,Weight)
plt.title('Correlation between Height and Weight')
plt.xlabel('Height (in cms)')
plt.ylabel('Weight (in cms)')
plt.show()
```

```python
calories = [65,75,95,100]
plt.plot(Weight,Height)
plt.plot(calories,Height)
plt.show()
```

```python
plt.plot(Weight,Height)
plt.plot(calories,Height)
plt.legend(labels = ['Weight','Calories'],loc ='lower right')

plt.ylabel('Height (in cms)')
plt.show()
```

```python
plt.plot(calories)
plt.plot(Weight)
plt.legend(labels = ['Weight','Calories'],loc ='lower right')

plt.ylabel('Height (in cms)')

plt.xticks(ticks = [0,1,2,3], labels =['p1','p2','p3','p4'])
plt.show()
```

```python
plt.figure(figsize = (10,7)) #width ,height
plt.plot(calories, 'k')
plt.plot(Weight,'--') # --,-.,-o.^,v,<,>
plt.legend(labels = ['Weight','Calories'],loc ='lower right')

plt.ylabel('Height (in cms)')

plt.xticks(ticks = [0,1,2,3], labels =['p1','p2','p3','p4'])
plt.show()
```

```python
#sub plots - Dividing the chart are into required pieces
fig , ax = plt.subplots(nrows = 2 , ncols = 2,figsize =(10,20))
ax[0,0].plot(calories, 'g-')
ax[0,1].plot(Weight, 'r-')


ax[0,0].set_title('calories')
ax[0,1].set_title('Weight')


```
