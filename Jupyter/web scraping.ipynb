```python
data = soup.find('table',{'class':"W(100%) M(0)"})
data = data.findAll('span')
data
```

```python
dat = data[7:-3]
dat = [x.text for x in dat]
dat = np.array(dat).reshape(100,7)
dat
```

```python
cols = data[:7]
cols = [col.text for col in cols]
cols
```

```python
df= pd.DataFrame(data = dat,column= cols)
print(df)
```
